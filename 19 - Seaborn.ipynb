# 19/30 Visualisation de Donn√©es avec Seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

## 1. Pairplot() : La vue d'ensemble

```python
sns.pairplot(iris, hue='species')
```

## 2. Visualiser de cat√©gories

```python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.set(font_scale=1, rc={'axes.edgecolor': 'black','axes.facecolor':'white','figure.facecolor':'white'})
sns.catplot(x='survived', y='age', data=titanic, hue='sex')
plt.show()
```

```python
plt.figure(figsize=(32, 8))
sns.boxplot(x='age', y='fare', data=titanic, hue='sex')
plt.show()
```

## 3. Visualisation de Distributions

```python
sns.distplot(titanic['fare'])
plt.show()
```

```python
sns.jointplot('age', 'fare', data=titanic, kind='hex')
plt.show()
```

```python
sns.heatmap(titanic.corr())
plt.show()
```

```python
values = np.random.normal(loc=3.0,scale=1.0,size=50000)
df_FLIM = pd.DataFrame(values, columns=['Values'])
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
sns.boxplot(df_FLIM["Values"], ax=ax_box, color='red')
sns.distplot(df_FLIM["Values"], ax=ax_hist, color='red')
ax_box.set(xlabel='')
plt.tight_layout()
plt.show()
```
